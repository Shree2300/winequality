```python
df.cols=['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
       'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
```

```python
df.columns=df.cols
df.columns
```

```python
# As mentioned in the content , i have considered quality 7 or greater than 7  as good to which i have assign the values 1.
# Similarly for quality less than 7 it has been considered not good to which i have assign those vales as 0.
```

```python
def tran_quality(x):
    if x == 3:
        return 0
    if x == 4:
        return 0
    if x == 5:
        return 0
    if x == 6:
        return 0
    if x == 7:
        return 1
    if x == 8:
        return 1
```

```python
df['quality']=df['quality'].apply(tran_quality)
df
```

```python
x=df.drop('quality',axis=1)
y=df['quality']
```

```python
x.skew()
```

```python
from sklearn.preprocessing import power_transform
x_new=power_transform(x)


x=pd.DataFrame(x_new,columns=x.columns)
```

```python
x.skew()
```

```python
#checking for outliers
#boxplot
x.boxplot(figsize=[15,8])
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split

maxAccu=0
maxRS=0
for i in range(1,1599):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    LR=LogisticRegression()
    LR.fit(x_train,y_train)
    predrf=LR.predict(x_test)
    acc=accuracy_score(y_test,predrf)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i

print('Best accuracy is ',maxAccu,'on random_state',maxRS)
```

```python
#creating train_test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=1458)

from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
predlr=LR.predict(x_test)
print('Accuracy',accuracy_score(y_test,predlr)*100)
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

# Decision tree classifier


```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
preddt=dt.predict(x_test)
print('Accuracy',accuracy_score(y_test,preddt)*100)
print(confusion_matrix(y_test,preddt))
print(classification_report(y_test,preddt))
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print('Accuracy',accuracy_score(y_test,predrf)*100)
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))

```

# SVC

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(x_train,y_train)
ad_pred=svc.predict(x_test)
print('Accuracy',accuracy_score(y_test,ad_pred)*100)
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

# Cross validation


```python
from sklearn.model_selection import cross_val_score

scr=cross_val_score(LR,x,y,cv=5)
print('Cross Validation score of Logistic Regression model :',scr.mean())
```
